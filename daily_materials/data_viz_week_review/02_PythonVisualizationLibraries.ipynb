<h2>Python visualization libraries</h2>
Every programming language has its strengths and weaknesses. Some languages like the R language have some very well established and easy to use plotting libraries like ggplot. One issue with Python is that there are so many different libraries and visualization tools available in Python. In the course, we will mostly focus on plotting using libraries based on matplotlib which is the oldest and by far the most popular Python plotting library. It mostly focuses on static images and offers a very extensive range of 2D plot types and output formats. You can customize just about anything you want in your plot. The problem is that using Matplotlib on its own can involve quite a bit of code.  You will find that a lot of libraries are built on top of matplotlib. In this course, we will primarily use matplotlib and various matplotlib wrappers such as the seaborn library and the plotting functionality of pandas library. These wrappers make it so you can spend less time on fine tuning every detail of a plot and more time on data analysis and machine learning.
