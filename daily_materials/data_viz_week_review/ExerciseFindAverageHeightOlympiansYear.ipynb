<h2>Understanding the dataset</h2>
There is a dataset on Kaggle called 120 years of Olympic history. This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016. This data was originally web scraped from www.sports-reference.com in May 2018 by a R programmer (the code used to <a href="https://github.com/rgriff23/Olympic_history/blob/master/R/olympics%20wrangle.R">scrape and organize the data</a> is on GitHub). Note that the Winter and Summer Games were held in the same year up until 1992. After that, they staggered them such that Winter Games occur on a four year cycle starting with 1994, then Summer in 1996, then Winter in 1998, and so on. A common mistake people make when analyzing this data is to assume that the Summer and Winter Games have always been staggered.

The dataset given to you <a href="https://s3.amazonaws.com/General_V88/boomyeah2015/codingdojo/curriculum/content/chapter/athleteEventsNoPersonal.csv">here (athleteEventsNoPersonal.csv)</a> is a randomly sampled subset of that dataset containing roughly 15% of the original people in the dataset (a person is given a unique ID). For the rows of the dataset below, it happens to be for a person who participated in multiple events for two different olympic games.
